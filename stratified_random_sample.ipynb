```python
import pandas as pd
from datetime import datetime
def conv_datetime(inputStr):
    print(' '.join(inputStr.split()[2:4]))
    date_time = datetime.strptime(' '.join(inputStr.split()[2:4]), '%H:%M:%S %d/%m/%Y')
    print(date_time)
    return date_time
def strat_random_sample(csv_path):
    '''
        Function to perform stratified random sampling
        :param csv_path: path to the csv file
        :type csv_path: str
        :return: bool 
    '''
    assert(isinstance(csv_path, str))
    try:
        df = pd.read_csv(csv_path)
        cols = [col for col in df.columns]
        df = df[(df['AudioMothCode'].notna()) &
                (df['FileSize'].notna()) &
                (df['Duration'].notna()) &
                (df['Comment'].notna()) &
                (df['FileSize'] >= 46080192) & # checking if clips are a minute long and exceed 46.1 MB
                (df['Duration'] >= 60)]
        df['hour'] = df['Comment'].apply(conv_datetime).dt.hour.tolist()
        df = df.groupby(['AudioMothCode', 'hour']).apply(lambda x: x.sample(1)).drop(columns='hour')
    
        if(df.size > 0):
            df.to_csv('stratified_random_sample.csv', index=False)
            return True
        return False
    except (FileNotFoundError):
        print('File Not Found')
    except Exception as e:
        print(e)
```

```python
filepath = 'Peru_2019_AudioMoth_Data_Full.csv'
strat_random_sample(filepath)
```
